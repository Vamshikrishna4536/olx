```python
ax=pd.crosstab(olx.Fuel_Type,olx.Transmission).plot(kind="bar",stacked=True,figsize=(25,10))
for i in ax.containers:ax.bar_label(i,color="red")
```

```python
# Data Aggregation or grouping - groupby()
# leftside of groupby must be numeric variable
# rightside of groupby within() must be categorical variable
# statistical function must be specified at the end. sum(),mean(),median(),std(),skew(),etc.

# what is the Average price of diesel,petrol and CNG Cars?
np.round(olx.Price.groupby(olx.Fuel_Type).mean(),2)
```

```python
# what is the avg/mean mil
np.round(olx.Mileage.groupby(olx.Fuel_Type).mean(),2)
```
